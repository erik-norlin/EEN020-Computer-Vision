2 Points in Homogeneous Coordinates.
\textit{Theoretical Exercise 1}

$
x = \begin{pmatrix} x_1 \\ x_2 \\ x_3 \end{pmatrix} \Rightarrow x = \begin{pmatrix} x_1/x_3 \\ x_2/x_3 \\ x_3/x_3 \end{pmatrix}
$ 

$
x^{(1)} = \begin{pmatrix} 4 \\ -16 \\ 2 \end{pmatrix} \Rightarrow x^{(1)} = \begin{pmatrix} 2 \\ -8 \\ 1 \end{pmatrix}
$ 

$
x^{(2)} = \begin{pmatrix} -3 \\ 7 \\ -1 \end{pmatrix} \Rightarrow x^{(2)} = \begin{pmatrix} 3 \\ -7/2 \\ 1 \end{pmatrix}
$ 

$
x^{(3)} = \begin{pmatrix} 9\lambda \\ 3\lambda \\ 6\lambda \end{pmatrix} \Rightarrow x^{(3)} = \begin{pmatrix} 3/2 \\ 1/2 \\ 1 \end{pmatrix}
$ 

```python
import numpy as np
from scipy.io import loadmat
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
def pflat(A):
    return A/A[-1,:]
```

```python
def mat_to_np(mat):    
    path = r'C:\Users\erikn\skola\EEN020-Computer-Vision\assignment-1\A1data\data\compEx1.mat'
    mat_data = loadmat(path)
    A = mat_data[mat]
    return np.array(A)
```

```python
def plot_points_2D(A):
    plt.scatter(A[0,:], A[1,:])
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.axis('equal')
    plt.savefig(r'C:\Users\erikn\skola\EEN020-Computer-Vision\assignment-1\report-images\x2D.png', dpi=300)
    plt.show()
```

```python
def plot_points_3D(A):
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.scatter(A[0,:], A[1,:], A[2,:], 'o')
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')
    ax.axis('equal')
    fig.savefig(r'C:\Users\erikn\skola\EEN020-Computer-Vision\assignment-1\report-images\x3D.png', dpi=300)
    plt.show()
```

```python
mat = 'x2D'
A = mat_to_np(mat)
A = pflat(A)
if mat == 'x2D':
    plot_points_2D(A)
elif mat == 'x3D':
    plot_points_3D(A)
print(A)
```
